```python
import librosa
import numpy as np
import IPython.display as ipd

from musa.plot import Plotter
plotter = Plotter()
```

```python
x, sr = librosa.load("../data/songs/song-1.mp3", sr=44100)
zoom = [50, 100] # seconds
```

```python
plotter.plot_signal(x, sr, title="Song", zoom=zoom);
```

```python
# Compute STFT
N, H = 1024, 512 # 23 ms, 12 ms
gamma = 10
X = librosa.stft(x, n_fft=N, win_length=N, hop_length=H, window="hann")
ts = np.arange(X.shape[1]) * (H / sr)
fs = np.arange(X.shape[0]) * (sr / N)

Y = np.log(1 + gamma * np.abs(X))
```

```python
plotter.plot_matrix(Y, ts, fs, title="Song-1", zoom=zoom);
```
